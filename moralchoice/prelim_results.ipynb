# Initial evaluation results

````python
def load_model_data(file_path, model_name):
    df = pd.read_csv(file_path)
    df['action1_likelihood'] = df.apply(lambda row: row['action_likelihood'] if row['decision'] == 'action1' else 1 - row['action_likelihood'], axis=1)
    df['model'] = model_name
    return df[['model', 'action1_likelihood']]

# Load data for multiple models
models_data = [
    load_model_data('results/high50_initial_claude-3.5-sonnet.csv', 'claude-3.5-sonnet'),
    load_model_data('results/high50_initial_gpt-4o.csv', 'gpt-4o'),
    load_model_data('results/high50_initial_claude-3-haiku.csv', 'claude-3-haiku'),
    load_model_data('results/high50_initial_gpt-4o-mini.csv', 'gpt-4o-mini'),
    load_model_data('results/high50_initial_llama-3.1-70b.csv', 'llama-3.1-70b'),
    load_model_data('results/high50_initial_mixtral-8x7b.csv', 'mixtral-8x7b'),
]

# Combine all model data
combined_df = pd.concat(models_data, ignore_index=True)

# Create the violin plot
sns.violinplot(x='model', y='action1_likelihood', data=combined_df)

# Customize the plot
plt.style.use('default') 
plt.title('Distribution of action likelihood (P(action1)) across models')
plt.xlabel('Model')
plt.ylabel('Action likelihood')
plt.xticks(rotation=45)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
````
